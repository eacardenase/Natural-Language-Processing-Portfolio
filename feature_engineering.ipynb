# Punto 1: Pre-Procesamiento

- `[5 pts]` Leer el archivo `dialogos.csv` usando `pandas`
- `[10 pts]` Crear una nueva columna con el texto en minúscula, sin caracteres especiales ni números y sin palabras vacias
- `[7 pts]` Usando la nueva columna, ¿cuáles son las 10 palabras más usadas? 

```python
cd ..
```

```python
texto = pd.read_csv("dialogos.csv")
```

```python
las = []

for column in texto:
    row = texto["Locución"]
    for palabra in row:
#         palabra.lower()
        las.append(palabra)
```

```python
las = [palabra.lower() for palabra in las]
```

```python
las
```

```python
limpio = []

for palabra in las:
    cleaned = re.sub(r"[\W\d]", " ", palabra)
    limpio.append(cleaned.strip())
```

```python
limpio
```

```python
limpio2 = []

for palabras in limpio:
    palabras = palabras.split()
#     for word in palabras:
#         if word not in stopwords_sp:
#             limpio2.append(word)
    limpio2.append(palabras)
```

```python
# limpio2
```

```python
texto_1 = limpio2[1]
```

```python
def listToString(lista):
    string = " "
    return string.join(lista)
```

```python
pre_procesado = []
for lista in limpio2:
    lista = [word for word in lista if word not in stopwords_sp]
    lista = listToString(lista)    
    pre_procesado.append(lista)
# pre_procesado
```

```python
pre_procesado_df = pd.DataFrame({'pre-procesado': pre_procesado})
```

```python
texto["pre-procesado"] = pre_procesado_df
```

```python
columna = texto["pre-procesado"]
```

```python
palabras = []

for palabra in columna:    
    lista = palabra.split()
    for word in lista:
        palabras.append(word)
    
# palabras
```

```python
conteo = pd.Series(palabras)
```

```python
conteo.value_counts().head(10)
```

- `[3pts]` ¿Cuándo es una mejor idea usar una bolsa de n-gramas en vez de una bolsa de palabras?

Una bolsa de n-gramas se usa cuando el orden y la sintaxis de las palabras son importantes. Por otro lado, las bolsas de palabras no tienen en cuenta este aspecto y pueden ignorar palabras como "Santa Fe de Bogota", creando una bolsa con las palabras ['Santa', 'Fe', 'de', 'Bogota'].
